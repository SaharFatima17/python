```python
import numpy as np

# probability calculator
def calc_probability(value):
    return 1 / (1 + np.exp(-value))

# email validation logic
def check_email_length(length):
    base = length - 10
    prob = calc_probability(base)

    result = "Valid Email" if prob >= 0.9 else "Invalid Email"
    return prob, result

# sample email lengths
email_lengths = [6, 19, 16, 12, 10, 14]

for length in email_lengths:
    prob, res = check_email_length(length)

    print(f"Email Length: {length}")
    print(f"Accuracy Probability: {prob:.2f}")
    print(f"Status: {res}")
    print("-" * 20)
```
